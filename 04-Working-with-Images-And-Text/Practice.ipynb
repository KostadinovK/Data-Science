```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

from skimage.io import imread,imshow
from scipy.ndimage.filters import convolve

from collections import Counter
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer

from sklearn.datasets import fetch_20newsgroups
```

```python
tiger_image = imread("http://www.abc.net.au/news/image/7318414-3x2-940x627.jpg")
imshow(tiger_image)
```

```python
red_channel = tiger_image[:,:,0]
green_channel = tiger_image[:,:,1]
blue_channel = tiger_image[:,:,2]
imshow(red_channel)
```

```python
imshow(green_channel)
```

```python
imshow(blue_channel)
```

```python
plt.hist(red_channel.ravel(),color="red",bins = 256,alpha = 0.5)
plt.hist(green_channel.ravel(),color="green",bins = 256, alpha = 0.5)
plt.hist(blue_channel.ravel(),color="blue",bins = 256, alpha = 0.5)
plt.show()
```

```python
average = tiger_image.mean(axis = 2)
imshow(average,cmap = "gray")
```

```python
tiger_grayscale = 0.299 * red_channel + 0.587 * green_channel + 0.114 * blue_channel
imshow(tiger_grayscale,cmap = "gray")
```

```python
box_blur_kernel = np.array([
    [1,1,1],
    [1,1,1],
    [1,1,1]]) / 9
blurred = convolve(tiger_grayscale,box_blur_kernel)
plt.imshow(tiger_grayscale[300:400,400:600],cmap = "gray")
plt.title("Normal")
plt.show()
plt.imshow(blurred[300:400,400:600],cmap = "gray")
plt.title("Blurred")
plt.show()
```

```python
with open("alice.txt","r") as file:
    alice = file.read()
print(len(alice))
```

```python
counter = Counter(alice)
counter.most_common(20)
```

```python
word_counter = Counter(re.split("\\W+",alice.lower()))
word_counter.most_common(20)
```

```python
alice_words = re.split("\\W+",alice.lower())
stop = stopwords.words("english")
counter_without_stopers = Counter([w for w in alice_words if w not in stop])
counter_without_stopers.most_common(20)
```

```python
stemmer = PorterStemmer()
alice_without_stopwords = [w for w in alice_words if w not in stop]
stemmer.stem(alice_without_stopwords[1])
```
